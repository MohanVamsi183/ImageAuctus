```python
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, Reshape, Flatten, Dropout, Input, BatchNormalization
from tensorflow.keras.layers import LeakyReLU
from tensorflow.keras.models import Sequential
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.datasets import cifar10

# Load and preprocess the CIFAR-10 dataset
(X_train, _), (_, _) = cifar10.load_data()
X_train = (X_train.astype(np.float32) - 127.5) / 127.5  # Normalize to [-1, 1]

# Set parameters
latent_dim = 100
adam = Adam(learning_rate=0.0002, beta_1=0.5)  # Increased learning rate

def build_generator():
    model = Sequential()
    model.add(Dense(256 * 8 * 8, input_dim=latent_dim))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())
    model.add(Reshape((8, 8, 256)))

    model.add(Conv2DTranspose(128, kernel_size=5, strides=2, padding='same'))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())

    model.add(Conv2DTranspose(64, kernel_size=5, strides=2, padding='same'))
    model.add(LeakyReLU(0.2))
    model.add(BatchNormalization())

    model.add(Conv2DTranspose(3, kernel_size=5, activation='tanh', padding='same'))
    return model

def build_discriminator():
    model = Sequential()
    model.add(Input(shape=(32, 32, 3)))
    model.add(Flatten())
    model.add(Dense(1024))
    model.add(LeakyReLU(0.2))
    model.add(Dropout(0.3))
    model.add(Dense(512))
    model.add(LeakyReLU(0.2))
    model.add(Dropout(0.3))
    model.add(Dense(1, activation='sigmoid'))
    return model

# Create the GAN
generator = build_generator()
discriminator = build_discriminator()
discriminator.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

# Combine models
discriminator.trainable = False
gan_input = Sequential()
gan_input.add(generator)
gan_input.add(discriminator)
gan_input.compile(loss='binary_crossentropy', optimizer=adam)

# Function to plot original images
def plot_real_images(examples=5, dim=(1, 5), figsize=(10, 2)):
    real_images = X_train[:examples]
    real_images = (real_images + 1) / 2.0  # Rescale original images for visualization

    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        plt.imshow(real_images[i])
        plt.axis('off')
        plt.title("Original Image")

    plt.tight_layout()
    plt.show()

# Function to plot generated images
def plot_generated_images(epoch, examples=5, dim=(1, 5), figsize=(10, 2)):
    noise = np.random.normal(0, 1, (examples, latent_dim))
    generated_images = generator.predict(noise)

    # Scale images back to [0, 1] for visualization
    generated_images = (generated_images + 1) / 2.0

    plt.figure(figsize=figsize)
    for i in range(examples):
        plt.subplot(dim[0], dim[1], i + 1)
        plt.imshow(generated_images[i])
        plt.axis('off')
        plt.title(f"Generated Image Epoch {epoch}")

    plt.tight_layout()
    plt.show()

# Training the GAN
def train_gan(epochs=1000, batch_size=128, save_interval=50):
    # Plot original images first
    print("Original Images:")
    plot_real_images()  # Show original images at the start

    for epoch in range(epochs):
        # Train the discriminator
        idx = np.random.randint(0, X_train.shape[0], batch_size)
        real_images = X_train[idx]
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        fake_images = generator.predict(noise)

        d_loss_real = discriminator.train_on_batch(real_images, np.ones((batch_size, 1)))
        d_loss_fake = discriminator.train_on_batch(fake_images, np.zeros((batch_size, 1)))

        # Train the generator
        noise = np.random.normal(0, 1, (batch_size, latent_dim))
        g_loss = gan_input.train_on_batch(noise, np.ones((batch_size, 1)))

        # Print progress every 50 epochs
        if epoch % 50 == 0:
            avg_d_loss = 0.5 * (d_loss_real[0] + d_loss_fake[0])
            avg_d_acc = 0.5 * (d_loss_real[1] + d_loss_fake[1])
            #print(f"Epoch: {epoch} | D Loss: {avg_d_loss:.4f} | D Acc.: {avg_d_acc:.4f} | G Loss: {g_loss:.4f}")

        # Plot generated images every 'save_interval' epochs and on the last epoch
        if epoch % save_interval == 0 or epoch == epochs - 1:
            print(f"Showing generated images for epoch {epoch}")
            plot_generated_images(epoch)

# Start training
train_gan(epochs=200, batch_size=128)
```
